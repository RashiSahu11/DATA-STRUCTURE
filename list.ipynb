LIST

```python
mob_app=['pinterest','whatsapp','telegram','instagram','swiggy','savana']
print(mob_app)

```

INDEXING AND SLICING

```python
print(mob_app[1])#pos indexing
print(mob_app[5])
```

```python
print(mob_app[-2])#neg indexing
print(mob_app[-5])
```

```python
#slicing
print(mob_app[1:5])#start,end
print(mob_app[1:5:1])#start:end:gap
print(mob_app[:5])#if start not given
print(mob_app[3:])#if end of lst not given
print(mob_app[:])#for whole list
```

LIST METHODS

- append -  adds an element at the end of the list 
- insert - adds an element at any specific position
- pop - delets an element from the end and also using index value
- remove - deletes an element using the name of it
- copy - make a copy of the list
- sort - sort the elements in specific order
- extend - combine two list

```python
mob_app.append('netflix')
print(mob_app)
```

```python
mob_app.insert(2,'maps')
print(mob_app)
```

```python
mob_app.pop() #delete last element by default
print(mob_app)
```

```python
mob_app.pop(0)
print(mob_app)
```

```python
mob_app.remove('maps')
print(mob_app)
```

```python
mob_app2=mob_app.copy()
print(mob_app2)
print(mob_app)
```

```python
mob_app2.clear() 
print(mob_app2)
```

```python
print(mob_app)
```

```python
mob_app.sort()
print(mob_app)
```

```python
lst=[1,2,3,4,5,6]
mob_app.extend(lst)
print(mob_app)
```

```python
lst.extend(mob_app)
print(lst)
```
