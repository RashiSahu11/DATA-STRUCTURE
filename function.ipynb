```python
def count_even(lst): # to count even numbers 
    count=0

    for num in lst:
        if num%2==0:
            count+=1
    return count

print(count_even([43,5,6,235,6,2,5,8,9]))    


```

```python
def count(lst): # to count odd numbers
    count=0

    for num in lst:
        if num%2!=0:
          count+=1
    return count
print(count([2,4,5,21,89,431,53]) )
```

```python
def counnt(lst): #to count odd numbers
    count=0

    for num in lst:
        if num%2==1:
           count+=1
    return count
print(counnt([45,56,23,432,676,43]))    
```
