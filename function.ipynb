`FUNCTION`: a block of code to reuse any functionality

`Types of functions`
- user defined functions
     - parametrised functions
     - non parametrised functions
     - return function
     - non return function
     - default parametrised functions
     - lambda functions 
- built in functions
     - append, len etc.        

```python
def greet(): #function definition
    print("hello user")
    print("welcome to code world")
    print("have a great day")
greet()    # function calling
print ("see u again")
```

```python
greet()
```

```python
#parametrised
def sum(a,b): #parameters
    print(a+b)
sum(10,20) #argument   
```

```python
#check no is even or odd
def evod(s):
    if s%2==0:
       print('even') 
    else:
       print('odd')    
evod(3)             
```

```python
# average of 4 numbers
def avg(a,b,c,d):
    sum=a+b+c+d
    print('average of numbers',sum/4)
avg(1,2,3,1)    
```

```python
#non return
def sum(a,b): 
    print(a+b)
result=sum(10,20) 
print(result)
```

```python
#return fuction
def sum(a,b):
    return a+b

result=sum(10,20)
print(result)
```

```python
#default parametrised function
def country(country='US'):
    print(f'I love {country}')

country('Belgium')
country()
```

```python
#take a list of 10 integers and return a new list with the square of all the elements
#take a list of 10 integers and return a new list of all the odd number
lst=[1,3,4,2,6,8,9,0,2,6]
lst2=[]
for i in lst:
    lst2.append(i**2)
print(lst2)
    

```

```python
lst=[1,3,4,2,6,8,0,2,6]
lst2=[]
for i in lst:
    if i%2!=0:
        lst2.append(i)
print(lst2)        
```

LAMBDA FUNCTION


variable=lambda parameter:expression

```python
square= lambda x:x**2
print(square(4))
```

```python
even = lambda x :'even' if x%2==0 else 'odd'
print(even(7))
```
