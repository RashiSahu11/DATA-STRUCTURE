LAMBDA FUNCTION


variable=lambda parameter:expression

```python
square= lambda x:x**2
print(square(4))
```

```python
even = lambda x :'even' if x%2==0 else 'odd'
print(even(7))
```

```python
#Write a function is_even(num) that returns True if the number is even and False if it is odd.
def check_no(num):
    if num%2==0:
        print("even")
    else:
        print("odd")    

print(check_no(9))        
```

```python
lst=[1,3,2,5,6,8]
squares = list(map(lambda x:x**2, lst))
print(squares)
```

```python
number=[1,2,3,4,5,6,7,8,9]
even_number=list(filter(lambda x:x%2==0, number))
print(even_number)
```
