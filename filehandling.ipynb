```python
#write mode
file=open("data.txt",'w')#open file in a write mode
file.write("This is a new line of text.\n")
file.close()
```

```python
#read mode
file=open('data.txt','r')
print(file.read())
file.close()
```

```python
#append mode
file=open('data.txt','a')
file.write("This is appended line")
file.close()
```

```python
#file automatically closes WITH
with open("data.txt",'r') as file:
    content=file.read()
    print(content)
```

```python
#read specific number of characters
file=open("data.txt",'r')
print(file.read(10))#read first 10 acharacters
file.close()
```

```python
#read one line at a time
file=open("data.txt",'r')
print(file.readline())#reads first line
file.close()
```

```python
file=open('data.txt','r')
print(file.name)#name of a file
print(file.mode)#mode in which the file is opened
print(file.closed)#check if the file is closed or not
print(file.read(5))#reads characters
print(file.tell())#currrent position of file cursor
file.seek(0)#move the file cursor to the beging of the file
```

EXCEPTION HANDLING


```python
age=int(input("Enter your age:"))
if age>18:
    print("Valid")
else:
    print("Invalid")    
```

```python
try:
    age=int(input("Enter your age:"))
    print(age)
except ValueError:  
    print("please enter a valid age")  
```

```python
try:
    file=open('newfile.txt','x')
    print("File created successfully")
except FileExistsError:
    print("File already exists")
finally:    
    file.close()        
```
