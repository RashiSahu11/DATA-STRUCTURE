DICTIONARY

 - different ways to create a dictionary
 - access key value pairs
 - adding a new pair
 - deleting a pair
 - updating a pair
 - loops in dictionary
 - nested dictionary
 - access value in nested dictionary

```python
student={'name':'rashi',
         'age':19,
        'subject':'python'
}        
print(student)
```

```python
temp={'banglore':[26,28,24],
      'chennai':[28,27,31],
      'uttrakhand':[28,24,26]
      }
print(temp)
```

```python
info = dict(name = 'joy', course='bca', age=20)
print(info)
```

```python
#accessing values using key
#1st method
student['name']
student['age']
```

```python
#2nd method
student.get('name')# get method
```

```python
print(student.keys())
print(student.values())
print(student.items())
```

```python
#adding a new pair
student['city']='banglore'
print(student)
```

```python
#deleting a pair
del student['city']
print(student)
```

```python
#update
student['subject']='data science'
print(student)
```

```python
#loops in dict
for i,v in student.items():
    print(f'{i}:{v}')
```

```python
a=10
b=20
print('sumis :',a+b)
print("sum of",a,'and',b,'is',a+b)
print(f'sum of {a} and {b} is {a+b}')#f string to format 
```

```python
#nested dictionary
company={
    'empoo1':{'name':'Rahul',
            'salary':50000,
            'department':'datascience'
    },
    'emp002':{'name':'raj',
              'salary':40000,
              'department':'powerbi'
    }
}
print(company)    

```

```python
company['emp003']= {
    'name':'sun',
    'salary':30000,
    'department':'AI'
}
print(company)                            
```

```python
company['emp002']['name']='raj yadav'
print(company)
```

```python
#calculate the avg tem of banglore (using the dictionary key and values pair)
temp={'banglore':[26,28,24],
      'chennai':[28,27,31],
      'uttrakhand':[28,24,26]
      }
avg=sum(temp['banglore'])/len(temp['banglore'])
print ('Average temperature of banglore is:',avg)
```
